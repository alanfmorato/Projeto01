```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import sql
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df_obesos = df[(df.Age > 50)]
df_obesos
```

```python
def Obesidade(BMI):
    if BMI > 30:
        return "1"
    elif BMI < 30:
        return "0"

df_obesos['Obesidade'] = df_obesos['BMI'].map(Obesidade)
display(df_obesos.head())
```

```python
df_obesos = df[(df_obesos.Obesidade == 1)]
df_obesos
```
