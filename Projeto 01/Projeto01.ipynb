```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
import sql
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df = df[(df.Age > 50)]
df
```

```python
def result(BMI):
    if BMI > 30:
        return 1
    elif BMI < 30:
        return 0

df['resultado'] = df['BMI'].map(result)
df.head()
```

```python
df = df[(df.resultado == 1)]
df.head()
```

```python
df.to_csv("tratado")
```
